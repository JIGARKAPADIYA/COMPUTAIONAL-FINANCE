Q.1 Brownian motion

```
import numpy as np
import matplotlib.pyplot as plt
import math as m
N=64
T=10
s=5

np.random.seed(s)
t=np.arange(0,N*T,1)
B=np.zeros(len(t))
for i in range(len(t)-1):
  B[i+1]=B[i]+np.random.normal()
plt.plot(t/N,B/np.sqrt(N)) 
plt.title("Brownian Motion  N=64")
plt.xlabel("Time (s)")
```

Q.2 Geometric Brownian motion

```
import numpy as np
import matplotlib.pyplot as plt
import math as m
N=64
T=2
s=5
u=0.15
sig=0.4
var=sig*sig
np.random.seed(s)
t=np.arange(0,N*T,1)
S=np.zeros(N*T)
for i in range(len(t)-1):
  B[i+1]=B[i]+np.random.normal()
S[0]=55.25
for i in range(N*T -1):
  S[i+1]=S[0]*m.exp(u*t[i]/N - var*t[i]/(2*N) + sig*B[i]/np.sqrt(N))
plt.plot(t/N,S)  
plt.xlabel("Time (s)")
plt.title("Geometric Brownian motion")
```

Q.3 AMZN stock price 

```
import csv
import pandas as pd
df = pd.read_csv('Amazon.csv')
adj = df['Adj. Close']
returns = np.zeros(len(adj)-1)
for i in range(1,len(adj)):
  returns[i-1]=(adj[i]-adj[i-1] )/ adj[i-1]
x=np.mean(returns)*252
print("mu is "+str(round(x,4)))
sig = np.sqrt(np.var(returns))*np.sqrt(252)
print("sigma is "+str(round(x,4)))
N=64
T=2
u=x
var=sig*sig
np.random.seed(22)
t=np.arange(0,N*T,1)
B=np.zeros(len(t))
S=np.zeros(len(t))
S[0]=636.99
for i in range(len(t)-1):
  B[i+1]=B[i]+np.random.normal()
for i in range(N*T -1):
  S[i+1]=S[0]*m.exp(u*t[i]/N - var*t[i]/(2*N) + sig*B[i]/np.sqrt(N))
plt.plot(t/N,S)  
plt.xlabel("Time (s)")
plt.title("Geometric Brownian motion")  
```
